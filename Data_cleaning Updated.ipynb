```python
#drop columns 
df_clean = df_clean.drop(['X','Y','OBJECTID','EVENT_UNIQUE_ID','LONG_WGS84','LAT_WGS84','HOOD_140','NEIGHBOURHOOD_140'],axis =1)
```

```python
# drop outliers
q3 = df_clean['Death and Injuries'].quantile(0.75)
q1 = df_clean['Death and Injuries'].quantile(0.25)
iqr = q3 - q1
upper_whisker = q3 + 1.5 * iqr
outlier = df_clean[df_clean['Death and Injuries'] > upper_whisker]
outlier
```

```python
#we need to drop the outlier
avg = df_clean['Death or Injuries'].mean()
std = df_clean['Death or Injuries'].std()
threshold = avg + 2 * std
df_clean['Death or Injuries'] = df_clean['Death or Injuries'] < threshold
```
